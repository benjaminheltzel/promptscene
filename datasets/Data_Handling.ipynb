````python
from glob import glob
import os
import shutil

ORIGINAL_REPLICA_PATH = os.path.join("data", "replica", "*")
PROCESSED_REPLICA_PATH = os.path.join("data", "replica_3d", "*")
SPLITTED_REPLICA_PATH = os.path.join("data", "replica_split")

# split train/val/test
train_test_dict = {
    "office0": "train",
    "office1": "train",
    "office2": "train",
    "office3": "val",
    "office4": "test",
    "room0": "train",
    "room1": "val",
    "room2": "test",
}
````

````python
os.makedirs(os.path.join(SPLITTED_REPLICA_PATH, "train"), exist_ok=True)
os.makedirs(os.path.join(SPLITTED_REPLICA_PATH, "val"), exist_ok=True)
os.makedirs(os.path.join(SPLITTED_REPLICA_PATH, "test"), exist_ok=True)

# Split the pre-processed data into train/val/test
for file_name in glob(PROCESSED_REPLICA_PATH):
    scene_name = os.path.splitext(os.path.basename(file_name))[0]
    print(f"Moving {file_name} -> {os.path.join(SPLITTED_REPLICA_PATH, train_test_dict[scene_name])}")
    shutil.move(file_name, os.path.join(SPLITTED_REPLICA_PATH, train_test_dict[scene_name]))

# Split the original data into train/val/test
for folder_name in glob(ORIGINAL_REPLICA_PATH):
    scene_name = os.path.basename(folder_name)
    if scene_name == "ground_truth":
        shutil.copytree(folder_name, os.path.join(SPLITTED_REPLICA_PATH, "ground_truth"))
        continue
    src_file = os.path.join(folder_name, f"{scene_name}_mesh.ply")
    dst_file = os.path.join(SPLITTED_REPLICA_PATH, train_test_dict[scene_name], f"{scene_name}_mesh.ply")
    print(f"Copying {src_file} -> {dst_file}")
    with open(src_file, 'rb') as src, open(dst_file, 'wb') as dst:
        shutil.copyfileobj(src, dst)

# remove replica_3d dir
os.rmdir(PROCESSED_REPLICA_PATH.replace('*', ''))
````

## Data Loading

````python
from point_loader import Point3DLoader 
point_loader = Point3DLoader(datapath_prefix="./data/replica_split",
                             voxel_size=0.02,
                             split='val', aug=False,
                             memcache_init=False,
                             eval_all=True,
                             input_color=False, 
                            )
````

#### result
- coords (OpenScene)
    - Represents the voxel positions in 3D space, with each entry having the form [1, x, y, z].
    - The shape is (N, 4), where N is the number of voxels.
- feats (OpenScene)
    - Color intensity for each voxel 
    - if input_color=False, all values are 1
    - Scaled to the range [-1, 1].
    - The shape is (N, 3), with 3 corresponding to RGB channels.
- labels (OpenScene)
    - all values are 255 because Replica dataset doesn't have semantic labels
- inds_reconstruct (OpenScene)
    - Provides indices for reconstructing the original position information of the voxels.
    - The shape is (N,), where N corresponds to the number of voxels, used to track voxel positions after voxelization.
- data (Mask3D)
    - data is a SparseTensor containing the following information
        - coordinates
            - A tensor of 3D coordinates (0, x, y, z) for the mesh vertices, stored in a sparse format
            - These coordinates represent the spatial positions of the vertices in the 3D space.
        - features
            - RGB colors associated with each coordinates
        - coordinate_map_key
            - A key used to manage the mapping between coordinates and their corresponding features in the sparse tensor
        - coordinate_manager
            - An object (CoordinateMapManagerCPU) that efficiently manages the storage and access of sparse coordinates. 
            - It uses an algorithm (MinkowskiAlgorithm.DEFAULT) to perform operations on the sparse tensor.
- points (Mask3D)
    - raw vertices
- colors (Mask3D)
    - The normalized color values for each raw vertices
- features (Mask3D)
    - The same data as colors, but specifically extracted for use as input features for the model
- unique_map (Mask3D)
    - An index map showing which coordinates are unique in the mesh, used to remove duplicates(?)
- inverse_map (Mask3D)
    - A reverse index map for converting back from unique coordinates to the original ones(?)
- point2segment (Mask3D)
    - point2segment=None
- point2segment_full (Mask3D)
    - point2segment_full=None

## Visualization
#### To use color information, change the Point3DLoader argument to input_color=True.

````python
import k3d
import numpy as np
def visualize_coords(coords, feats, point_size=1.0):
    points = coords[:, 1:].numpy()
    colors = (feats.numpy() + 1) * 127.5
    colors = colors.astype(np.uint64) 
    colors_hex = (colors[:, 0] << 16) + (colors[:, 1] << 8) + colors[:, 2] # convert  to 0xRRGGBB format
    
    plot = k3d.plot()
    point_cloud = k3d.points(positions=points,  point_size=point_size, colors=colors_hex)
    plot += point_cloud

    # display
    plot.display()
sample_data = point_loader[0]
visualize_coords(sample_data['coords'], sample_data['feats'])  # To use color information, change the PointLoader argument to input_color=True.
````
